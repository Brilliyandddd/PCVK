```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
print('Mengubah tingkat kecerahan citra')
print('--------------------------------')

try:
  brightness = int(input('Masukkan nilai brightness : '))
except ValueError:
  print("Error, Not a number.")

original = cv.imread('/content/drive/MyDrive/PCVK/Week3/female.tiff')
brightness_image = np.zeros(original.shape, original.dtype)

#akses pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c]=np.clip(original[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((original, brightness_image))
# cv2_imshow(original)
cv2_imshow(final_frame)
```

## Inverse Color

```python
original = cv.imread('/content/drive/MyDrive/PCVK/Week3/female.tiff')
inverse_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      inverse_image[y,x,c] =np.clip(255 - original[y,x,c], 0, 255)

final_frame2=cv.hconcat((original, inverse_image))
cv2_imshow(final_frame2)
```

## Contrast

```python
print('Mengubah tingkat kontras citra')
print('--------------------------------')

try:
  contrast = int(input('Masukkan nilai contrast : '))
  brightness = int(input('Masukkan nilai brightness : '))
except ValueError:
  print("Error, Not a number.")

original = cv.imread('/content/drive/MyDrive/PCVK/Week3/female.tiff')
kontras_image = np.zeros(original.shape, original.dtype)

#akses pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      kontras_image[y,x,c]=np.clip(contrast * original[y,x,c] + brightness, 0, 255)

final_frame3 = cv.hconcat((original, brightness_image, kontras_image))
# cv2_imshow(original)
cv2_imshow(final_frame3)
```

## Logarithmic

```python
print('Logarithmic citra')
print('--------------------------------')

try:
  const = int(input('Masukkan nilai konstanta : '))
except ValueError:
  print("Error, Not a number.")

original = cv.imread('/content/drive/MyDrive/PCVK/Week3/female.tiff')
log_image = np.zeros(original.shape, original.dtype)

#akses pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      log_image[y,x,c]=np.clip(const * (np.log(1 + original[y,x,c])),0 ,255)

final_frame = cv.hconcat((original, log_image))
cv2_imshow(final_frame)
```

## Grayscale

### Average

```python
print('Grayscale average citra')
print('--------------------------------')

original = cv.imread('/content/drive/MyDrive/PCVK/Week3/female.tiff')
gsavr_image = np.zeros(original.shape, original.dtype)

# original = cv.cvtColor(original, cv.COLOR_BGR2GRAY)

#akses pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      red = original[y, x, 0]
      green = original[y, x, 1]
      blue = original[y, x, 2]
      avg = (red + green + blue) / 3
      gsavr_image[y,x,c] = np.clip(avg, 0, 255)

final_frame = cv.hconcat((original, gsavr_image))
cv2_imshow(final_frame)
```

### Lightness

```python
print('Grayscale Lightness citra')
print('--------------------------------')

original = cv.imread('/content/drive/MyDrive/PCVK/Week3/female.tiff')
gslgh_image = np.zeros(original.shape, original.dtype)

#akses pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      red = original[y, x, 2]
      green = original[y, x, 1]
      blue = original[y, x, 0]
      maxmin = max(red, green, blue) + min(red, green, blue)
      light = maxmin / 2
      gslgh_image[y,x,c] = np.clip(light, 0, 255)

final_frame = cv.hconcat((original, gslgh_image))
cv2_imshow(final_frame)
```

### Luminance

```python
print('Grayscale Luminance citra')
print('--------------------------------')

original = cv.imread('/content/drive/MyDrive/PCVK/Week3/female.tiff')
gslum_image = np.zeros(original.shape, original.dtype)

#akses pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      red = original[y, x, 2]
      green = original[y, x, 1]
      blue = original[y, x, 0]
      red2 = 0.21 * red
      green2 = 0.72 * green
      blue2 = 0.07 * blue
      lumin = red2 + green2 + blue2
      gslum_image[y,x,c] = np.clip(lumin, 0, 255)

final_frame = cv.hconcat((original, gslum_image))
cv2_imshow(final_frame)
```
