### Gamma Correction

```python
import cv2
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/drive/MyDrive/PCVK/Week5/female.tiff')
```

```python
image_gamma = np.power(image / 255.0 , 1 / gamma)
image_gamma = np.uint8(255 * image_gamma)
```

### Image Depth

```python
import matplotlib.pyplot as plt

original = cv2.imread('/content/drive/MyDrive/PCVK/Week5/female.tiff', cv.IMREAD_GRAYSCALE)
depth_image = np.zeros(original.shape, original.dtype)

bit_depth = 2
level = 255 / (pow(2, bit_depth) -1)

for i in range(original.shape[0]):
  for j in range(original.shape[1]):
    depth_image[i][j] = round(original[i][j] / level) * level

plt.imshow(np.asarray(original), cmap='gray')
plt.imshow(np.asarray(depth_image), cmap='gray')

final_frame = cv.hconcat((original, depth_image))
cv2_imshow(final_frame)
```

### Average Denoising

```python
import glob

cv_img = []
for img in glob.glob('/content/drive/MyDrive/PCVK/Week5/noises/5.jpg'):
 n= cv.imread(img)
 cv_img.append(n)

def average_denoising(img):
    kernel = np.ones((5,5),np.float32)/25

    dst = cv.filter2D(img,-1,kernel)

    psnr = cv.PSNR(img, dst)

    return dst, psnr

for i in range(len(cv_img)):
    denoised_img, psnr = average_denoising(cv_img[i])
    print(f"PSNR of image {i+1}: {psnr:.2f}")
    cv2_imshow(cv_img[i])
    cv2_imshow(denoised_img)
```
